```python
import os
import cv2
import matplotlib.pyplot as plt
import pytesseract
import easyocr



def read_images_from_folder(folder_path):
    image_files = [f for f in os.listdir(folder_path) if f.endswith(('.jpg', '.jpeg', '.png'))]
    return [os.path.join(folder_path, img) for img in image_files]

def display_images_with_text(image_folder):
    image_paths = read_images_from_folder(image_folder)

    reader = easyocr.Reader(['en'])

    for image_path in image_paths:
        image = cv2.imread(image_path)

        # Use EasyOCR to extract text
        results = reader.readtext(image)

        # Display the image using plt
        plt.figure(figsize=(8, 8))
        plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        plt.title(f"Image: {os.path.basename(image_path)}")
        plt.axis('off')
        plt.show()


        # Print EasyOCR text
        print("EasyOCR Text:")
#         text = ""
#         for result in results:
#             text = ''.join(result[1])
            
#         print(text)

        # Extract text and concatenate into a single string
        concatenated_text = ' '.join([item[1] for item in results])
        print(concatenated_text)

        print("\n" + "=" * 40 + "\n")  # Separate images for better visibility
```

```python
# Example usage
cropped_images_folder = "cropped_images"
display_images_with_text(cropped_images_folder)
```
